```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_excel('/content/Groceries_Sales_data.xlsx')
```

```python
# Convert Date column
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Sort by date
data = data.sort_values('Date')
```

```python
#  Basic Info
print(data.head())
print(data.info())
```

```python
#  Sales Trend (Daily)
plt.figure()
plt.plot(data['Date'], data['Sales'])
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
#  Monthly Sales Analysis
data['Month'] = data['Date'].dt.month
monthly_sales = data.groupby('Month')['Sales'].sum()
monthly_sales.plot(kind='bar')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Yearly Sales Analysis
data['Year'] = data['Date'].dt.year
yearly_sales = data.groupby('Year')['Sales'].sum()

yearly_sales.plot(kind='bar')
plt.title('Yearly Sales Comparison')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
#  Best Sales Day
best_day = data.loc[data['Sales'].idxmax()]
print("\nBest Sales Day:")
print(best_day)
```

# Key Insights:
Sales show increasing/decreasing trend over time
Certain months have higher sales (seasonal pattern)
One specific day had maximum sales
Yearly comparison shows business growth/decline
